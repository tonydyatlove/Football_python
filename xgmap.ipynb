```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('xgtutorial.csv', on_bad_lines='skip')
```

```python
df.head()
```

```python
df
```

```python
a_xG = [0]
h_xG = [0]
a_min = [0]
h_min = [0]

#find home and away teams

hteam = df['team'].iloc[0]
ateam = df['team'].iloc[-1]

for x in range(len(df['xG'])):
    if df['team'][x]==ateam:
        a_xG.append(df['xG'][x])
        a_min.append(df['minute'][x])
    if df['team'][x]==hteam: 
        h_xG.append(df['xG'][x])
        h_min.append(df['minute'][x])
```

```python
a_xG
```

```python
def nums_cumulative_sum(nums_list):
    return [sum(nums_list[:i+1]) for i in range(len(nums_list))]
a_cumulative = nums_cumulative_sum(a_xG)
h_cumulative = nums_cumulative_sum(h_xG)
```

```python
a_cumulative
```

```python
fig, ax = plt.subplots(figsize = (10, 5))
fig.set_facecolor('#3d4849')
ax.patch.set_facecolor('#3d4849')

plt.xticks([0, 15, 30, 45, 60, 75, 90])
plt.xlabel('Minute')
plt.ylabel('xG')

ax.step(x=a_min, y=a_cumulative)
ax.step(x=h_min, y=h_cumulative)
```
